```python
import numpy as np
from scipy.stats import multivariate_normal

import matplotlib.pyplot as plt


mean = [0,0,0]
cov = [[10, 0, 0],
       [0, 10, 5],
       [0, 5, 10]]

out= np.random.multivariate_normal(mean, cov, 50000)
x,y,z = out[:,0],out[:,1],out[:,2]
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x,y,z, alpha = 0.1)
```
